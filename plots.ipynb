```python
import pandas as pd

import matplotlib.pyplot as plt

# Create the data
data = {
    'Class': ['Person', 'Car', 'Bicycle'],
    'GT': [355, 267, 86],
    'Det': [2611, 1339, 796],
    'Precision': [0.886, 0.915, 0.835],
    'Recall': [0.822, 0.954, 0.81],
    'AP': [0.901, 0.968, 0.884],
    'mAP@50': [0.475, 0.724, 0.549]
}

# Create DataFrame
df_map50 = pd.DataFrame(data)

# Display the table
print(df_map50)

# Bar plot for mAP@50
plt.figure(figsize=(6,4))
plt.bar(df_map50['Class'], df_map50['mAP@50'], color=['skyblue', 'orange', 'green'])
plt.xlabel('Class')
plt.ylabel('mAP@50')
plt.title('mAP@50 by Class')
plt.ylim(0, 1)
plt.show()
```
